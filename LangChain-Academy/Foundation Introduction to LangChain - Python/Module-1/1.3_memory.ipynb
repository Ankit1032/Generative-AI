## No memory

```python
from langchain.agents import create_agent


agent = create_agent(
    "gpt-4.1-nano"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    "gpt-5-nano",
    checkpointer=InMemorySaver(),  
)
```

```python
'''
We define our thread IDs so that we group these checkpoints of our state together
'''

from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Ankit and my favourite colour is Red")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```

By default, our agent state tracks a list of messages only, but we can add custom fields like user id, language, for example, if we would like that to be tracked over time.
