### Tools

##### Custom Weather Tool

```python
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"
```

##### Custom Reverse Text Tool

```python
def reverse_text(text: str) -> str:
    """Reverse any text."""
    return text[::-1]
```

##### Custom Calculate Expression Tool

```python
def calculate(expression: str) -> str:
    """Solve a math expression."""
    try:
        result = eval(expression)
        return str(result)
    except:
        return "Invalid expression"
```

#### Creating Agent with GPT 4o

```python
agent = create_agent(
    model="gpt-4o-mini",
    tools=[get_weather, reverse_text, calculate],
    system_prompt="You are a helpful assistant",
)
```

##### [Optional] If you want to change or fine-tune the model parameters like temperature, top_p, etc.

```python
#### If you want to change or fine-tune the model parameters like temperature, top_p, etc.
from langchain_openai import ChatOpenAI
model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.1,
    max_tokens=1000,
    timeout=30
    # ... (other params)
)

agent = create_agent(model, tools=[get_weather, reverse_text, calculate])
```

##### Prompting by testing the 1st Tool (Weather Tool)

```python
question = "what is the weather in sf"
```

```python
# Run the agent
response = agent.invoke(
    {"messages": [{"role": "user", "content": question}]}
)
```
