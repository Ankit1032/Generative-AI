```python
from dotenv import load_dotenv
import os
from langchain_openai import ChatOpenAI
```

```python
# Load env file
load_dotenv()

# Read API key
api_key = os.getenv("OPENAI_API_KEY")
```

```python
print("Loaded API key:", "YES" if api_key else "NO")
```

```python
# Create LLM client
llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=api_key,
)
```

```python
# Ask a question
response = llm.invoke("What is full form of OOO in corporate. Only give 1 sentence answer")
```
