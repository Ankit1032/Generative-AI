#### Criticism of Prompt Engineering:

1. Can change due to LLM Change. New LLMs can make the old LLMs obsolete
2. Depends on previous modules to work correctly
3. Manual supervision needed
4. Smarter LLMs may not need overtuned prompts. In future, with new smarter LLMs, we won't have to write ultra specific prompts or give too many few shot examples.

Instead of over-engineering prompts, we define how the data flows through different modules within our LLM program.

DSPy imagines every LLM as a text transformation graph.
Each node in this graph receives one or more text inputs, applies some text transformation and then passes it onto the next node for further processing.

Until the very end, we receive our desired output

DSPy completely detaches the underlying model and the program you designed meaning you can train both LLM and SLM to perform the same task.
