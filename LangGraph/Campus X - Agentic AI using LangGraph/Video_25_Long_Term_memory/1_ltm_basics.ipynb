```python
from langgraph.store.memory import InMemoryStore
```

```python
# creating a store
store = InMemoryStore()
```

```python
# creating a namespace
namespace = ("user", "u1") #to understand this, think u1 is a folder inside user and we will store the memory of u1 in u1
# I can also create "profile" inside u1 which will store the name of the user, age, etc.
# We can create unlimited profile to segment the what data we want to store like profile, preferences, etc.
# Namespace are a way to organise memory inside memory store.
```

if memory store is a google drive then namespace is a folder inside google drive.

### Creating Memories

1. "put" method inserts a new memory in namespace.

2. "put" requires 3 parameters :

    a. namespace : Where it wants to store the memory

    b. key: Unique key of that memory
    
    c. Value: Value inside the memory

```python
# Adding memories
store.put(namespace, "1", {"data": "User likes pizza"})
store.put(namespace, "2", {"data": "User prefers dark mode"})
```

```python
# another namespace for 2nd user
namespace2 = ("user", "u2")

# Add memories
store.put(namespace2, "1", {"data": "User likes pasta"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```

### Retrieving Memories

"get" method is used to retrieve memory using namespace name and the unique key

```python
# store.get(namespace, key)
store.get(namespace2, "2")
```

### Retrieving all Memories

```python
items = store.search(namespace2)

for item in items:
    print(item.value)
```

Now lets say your memory contains 100s of memories and you want to search some particular memory/details of the user. We cannot use "get" nor "search" as they return user wise memory or all namespace memory. So for this situation, we need to use semantic search.

### Semantic Search

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
# We are saying this particular memory store has capability to perform semantic search
# as i am passing embedding model into it.
store = InMemoryStore(index={'embed': embedding_model, 'dims':1536})
```

```python
namespace = ('users', 'u1')
```

```python
store.put(namespace, "1", {"data": "User prefers concise answers over long explanations"})
store.put(namespace, "2", {"data": "User likes examples in Python"})
store.put(namespace, "3", {"data": "User usually works late at night"})
store.put(namespace, "4", {"data": "User prefers dark mode in applications"})
store.put(namespace, "5", {"data": "User is learning machine learning"})
store.put(namespace, "6", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace, "7", {"data": "User prefers step-by-step reasoning"})
store.put(namespace, "8", {"data": "User is based in India"})
store.put(namespace, "9", {"data": "User likes real-world analogies"})
store.put(namespace, "10", {"data": "User prefers bullet points over paragraphs"})
```

```python
# we are searching some details about user though semantic search.
# limit means how many memory are supposed to be returned.
items = store.search(namespace, query="what is the user currently learning", limit=1)

for item in items:
    print(item.value)
```

```python
items = store.search(namespace, query="what are user's preferences", limit=3)

for item in items:
    print(item.value)
```
